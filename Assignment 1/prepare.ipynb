####Convert the text file to CSV file

```python
import csv

# File paths
input_file = "/content/SMSSpamCollection"
output_file = "SMSSpamCollection.csv"

with open(input_file, 'r', encoding='utf-8') as infile, open(output_file, 'w', encoding='utf-8', newline='') as outfile:
    writer = csv.writer(outfile)
    writer.writerow(['label', 'message'])

    for line in infile:
        label, message = line.strip().split('\t', 1)
        writer.writerow([label, message])

print(f"Conversion complete: {output_file} created.")
```

####Import necessary libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import os
from sklearn.preprocessing import LabelEncoder
```

####Utility Functions

```python
# 1.Function to load data
def load_data(file_path):
    data = pd.read_csv(file_path, encoding='utf-8')
    return data


# 2.Function for preprocessing data
def preprocess_data(data):
    label_encoder = LabelEncoder()

    # Encode the 'label' column ( 'ham' as 0 and 'spam' as 1)
    data['label'] = label_encoder.fit_transform(data['label'])
    # Remove leading/trailing spaces in messages
    if data['message'].dtype == 'object':
        data['message'] = data['message'].str.strip()
    return data


# 3.For splitting the dataset
def split_data(data, train_size=0.7, validation_size=0.15, test_size=0.15, random_state=42):

    # Ensure sizes add up to 1
    assert train_size + validation_size + test_size == 1, "Sizes must sum to 1"

    # Initial split into train and temp (validation + test)
    train_data, temp_data = train_test_split(data, test_size=(validation_size + test_size), random_state=random_state)
    # Split temp data into validation and test
    validation_data, test_data = train_test_split(temp_data, test_size=(test_size / (validation_size + test_size)), random_state=random_state)

    return train_data, validation_data, test_data



# 4.Function to save the splitted data
def save_splits(train_data, validation_data, test_data, output_dir='.', train_file='train.csv', validation_file='validation.csv', test_file='test.csv'):

    # Ensure the output directory exists
    os.makedirs(output_dir, exist_ok=True)

    # Save each split to a CSV file
    train_data.to_csv(os.path.join(output_dir, train_file), index=False)
    validation_data.to_csv(os.path.join(output_dir, validation_file), index=False)
    test_data.to_csv(os.path.join(output_dir, test_file), index=False)


```

```python
if __name__ == "__main__":
    # Load the dataset
    file_path = "/content/SMSSpamCollection.csv"
    data = load_data(file_path)

    # Preprocess the dataset
    data = preprocess_data(data)

    # Split the dataset
    train_data, validation_data, test_data = split_data(data)

    # Save the splits
    save_splits(train_data, validation_data, test_data, output_dir='.')
    print("Data preparation complete. Files saved to current directory.")
```
